```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"G:\ProdigyInfotech\Prodigy_Task_2\Titanic.csv")
df.head()
```

```python
#check for missing or null Values
print(df.isnull().sum())
```

```python
# Handling missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
# Explore the distribution of numerical features
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Explore the distribution of categorical features
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
# Explore the relationship between variables
sns.scatterplot(x='Age', y='Fare', data=df, hue='Survived', palette='Set1')
plt.title('Age and Fare by Survival')
plt.show()
```

```python
# Explore survival rates across different categories
sns.barplot(x='Pclass', y='Survived', data=df, hue='Sex', palette='viridis')
plt.title('Survival Rates of Passenger Class and Gender')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
